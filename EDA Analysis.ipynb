# Step3 : Univariate Analysis -> PDF, Histograms, Boxplots, Countplots, etc.

```python
numerical_columns = df.select_dtypes(include=['number']).columns
print(numerical_columns)
```

```python
#import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Univariate analysis for each numerical column
for col in numerical_columns:
    print(f"Univariate analysis for {col}")
    
    print(df[col].describe())
    
    # Visualization: Histogram
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

    # Visualization: Boxplot (to detect outliers)
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

# Observation

```python
1. Numerical Columns:

Salary Distribution: There might be a skewed distribution in the salary data (right-skewed), indicating that a few individuals have much higher salaries than others.
Academic Scores: The distributions of 10percentage, 12percentage, and collegeGPA could highlight how most individuals perform in the respective academic stages. For example, a normal distribution in these variables would show that most students scored around the mean, while skewness might indicate performance differences.
Skill Scores (English, Logical, Quant, etc.): Skill scores like English, Logical, Quant might show patterns of proficiency. Peaks in certain score ranges could suggest strengths in particular skill areas.
Personality Traits (Conscientiousness, Agreeableness, etc.): The distributions of the Big Five personality traits might reveal whether these traits are evenly distributed in the population or if there are tendencies toward specific traits, such as higher conscientiousness in this dataset
```

# Step 4: Bivariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot between 'Salary' and 'collegeGPA'
plt.figure(figsize=(10, 6))
for col in numerical_columns:
    sns.scatterplot(x=df[col], y=df['Salary'])
    plt.title(f'Scatter Plot of Salary vs {col}')
    plt.xlabel(col)
    plt.ylabel('Salary')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
for col in numerical_columns:
    plt.hexbin(df[col], df['Salary'], gridsize=30, cmap='Blues')
    plt.colorbar(label='Count')
    plt.title(f'Hexbin Plot of Salary vs {col}')
    plt.xlabel(col)
    plt.ylabel('Salary')
    plt.show()
```

```python
# List categorical columns specifically with 'category' dtype
categorical_columns = df.select_dtypes(include=['category']).columns
print("Categorical columns with 'category' dtype:", categorical_columns)
```

```python
# List categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print("Categorical columns:", categorical_columns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cross-tabulation between 'Gender' and 'Degree'
cross_tab_gender_degree = pd.crosstab(df['Degree'], df['Gender'])

# Plot stacked bar plot
cross_tab_gender_degree.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='coolwarm')
plt.title('Stacked Bar Plot of Degree vs Gender')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
# Cross-tabulation between 'Specialization' and 'CollegeState'
cross_tab_specialization_state = pd.crosstab(df['CollegeState'], df['Specialization'])

# Plot stacked bar plot
cross_tab_specialization_state.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title('Stacked Bar Plot of College State vs Specialization')
plt.xlabel('College State')
plt.ylabel('Count')
plt.legend(title='Specialization', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Cross-tabulation between 'JobCity' and 'Gender'
cross_tab_jobcity_gender = pd.crosstab(df['JobCity'], df['Gender'])

# Plot stacked bar plot
cross_tab_jobcity_gender.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='plasma')
plt.title('Stacked Bar Plot of Job City vs Gender')
plt.xlabel('Job City')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

# Observations on Bivariate Analysis

```python
Conclusion of Bivariate Analysis:
1. Relationships between Numerical Columns:

Salary vs. Academic Scores: Scatter plots or correlation analysis between Salary and collegeGPA, 10percentage, and 12percentage might show that higher academic performance correlates with better salaries. However, the strength of this relationship could vary depending on the column.
Salary vs. Skill Scores (English, Logical, Quant, etc.): Scatter plots could reveal that individuals with higher scores in these skill areas tend to have higher salaries. For example, stronger quantitative and logical skills might be closely associated with higher-paying jobs.
Salary vs. Personality Traits: Relationships between Salary and personality traits like conscientiousness or agreeableness could reveal how certain traits influence earning potential. For instance, higher conscientiousness might correlate with higher salary due to a stronger work ethic.
2. Relationships between Categorical and Numerical Columns:

Gender vs. Salary: Box plots could show that one gender tends to earn more than the other, indicating potential wage gaps.
Specialization vs. Salary: Some specializations might offer higher earning potential than others. For example, Engineering or Computer Science degrees could have higher average salaries compared to other fields like Arts or Management.
Degree vs. Salary: Higher-level degrees (Master’s, Ph.D.) could correlate with higher salaries compared to Bachelor’s degrees.
3. Relationships between Categorical Columns:

Gender vs. Specialization: A stacked bar plot or Chi-square test might reveal that certain specializations are more commonly chosen by one gender. For example, males might dominate in Engineering, while females could be more represented in Arts or Management.
JobCity vs. CollegeState: A stacked bar plot could reveal a trend where students from specific states prefer to work in certain cities, indicating regional preferences in job placements
```
